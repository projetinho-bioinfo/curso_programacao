## 1
Reescrever as funções dos dois exercícios da lista anterior que lidam com fasta, mas dessa vez usando biopython:

```python
from Bio import SeqIO

def conta_multifasta(arquivo_fasta):
  contador = 0
  for sequencia in SeqIO.parse(arquivo_fasta, 'fasta'):
    contador += 1
  return contador

print(conta_multifasta("./arquivos/multifasta.fasta"))

def tamanho_seq(arquivo_fasta):
  return len(SeqIO.read(arquivo_fasta, 'fasta')) # read lê apenas fastas com uma sequencia 

print(tamanho_seq("./arquivos/seq1.fasta"))
```

## 2
Escrever um script que converta um arquivo genbank em um arquivo fasta:

```python
def gb_to_fa(genbank):
  return SeqIO.read(genbank, 'genbank').format('fasta')

print(gb_to_fa("./arquivos/pgracilis_mitocondria.gb"))
```
